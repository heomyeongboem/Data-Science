#실습과제
이번에는 어도비 (Adobe)의 직원 분포를 한번 살펴봅시다.

어도비 전체 직원들의 직군 분포를 파이 그래프로 그려보세요.

(인원이 0인 직군은 그래프에 표시되지 않아야 합니다.)

```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('silicon_valley_details.csv')

#조건식
#회사 는 Adobe
c1 = df['company'] == 'Adobe'
#race 는 Overall_totals
c2 = df['race'] == 'Overall_totals'
#count 0인거 제외
c3 = df['count'] != 0
#job_category 가 Totals 랑 Previous_totals인거 제외
c4 = (df['job_category'] != 'Totals') & (df['job_category'] != "Previous_totals")

#조건식 합치기 
ans = df[c1 & c2 & c3 & c4]
#job_category를 기준으로 
ans.set_index('job_category',inplace= True)
#count를 세겠다. 
ans.plot(kind= "pie", y= "count")
```

##평균값: 잘못됫거나 특이한 데이터에 대해서 영향을 받음 
예)1 2 3 4 5 6 7 10000  -> mean = 5000??

장점: 특이한 데이터가 없을때는 더 좋은 insight를 가져옴 
## 중간값: 잘못됫거나 특이한 데이터에 대해서 영향을 안받음 
예)1 2 3 4 5 6 7 10000  -> median =  4 + 5// 2 = 4.5

짝수일때는 ((len/2 - 1) + len/2 )//2  예)8 --> 4 5 //2 

홀 수 일때는 len/2 + 1  예)9 --> 5 
단점: 특이한 데이터가 없을때는 안 좋은 insight를 가져옴 

```python
x = 70000 + 54000 + 97000 + 83000 + 66000 + 80000 
print(x)
y = 77000 * 7 - x 
print(y)
```

groupby 문법을 사용해서 '여성 비율'이 높은 순으로 직업을 나열해 보세요.

DataFrame이 아닌 Series로, 'gender'에 대한 값만 아래와 같이 출력되어야 합니다.

```python
#버젼 문제 
import pandas as pd

df = pd.read_csv('occupation.csv')

occupation_group = df.groupby('occupation')

df.loc[df['gender'] == 'M', 'gender'] = 0 # M이면 -> 0
df.loc[df['gender'] == 'F', 'gender'] = 1 # F이면 -> 1로 원핫 인코딩 하면 

occupation_group.mean()['gender'].sort_values(ascending=False) # 평균시 0.25면 여성 비율이 25%를 의미 
```
